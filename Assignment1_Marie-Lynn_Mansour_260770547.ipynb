**1. Filter the noisy image using a 5x5 Gaussian with variance equal to 2 (3 points)**

```
noisy_lighthouse = cv2.imread('/gaussian_lighthouse.png') # reading the image

original = cv2.cvtColor(noisy_lighthouse, cv2.COLOR_BGR2GRAY) # converting to grayscale
height, width = original.shape
gaussian = cv2.cvtColor(noisy_lighthouse, cv2.COLOR_BGR2GRAY)
height, width = gaussian.shape

from math import sqrt
gaussian_image = cv2.GaussianBlur(gaussian, (5,5), sqrt(2)) # 5x5 gaussian filter

psnr_values = cv2.PSNR(original, gaussian_image) # obtaining the psnr values

# plotting the figures
plt.figure(figsize = (18,18))
plt.subplot(131), plt.imshow(original, cmap = "gray")
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(gaussian_image, cmap = "gray")
plt.title("Gaussian 5x5 Filter\nPSNR " + str(psnr_values)), plt.xticks([]), plt.yticks([])
plt.show()
```

**Apply a Sobel edge detector with the kernel size of 3x3, 5x5 and 7x7 to the image. Threshold the filtered image to detect edges. Use two values of thresholds: 10% and 20% of the maximum pixel value in the filtered image. (4 points)**

```
cameraman = cv2.imread('/cameraman.png', cv2.IMREAD_GRAYSCALE) # reading the image
```

3 x 3 Kernel size

```
cameraman_sobel_x = cv2.Sobel(cameraman, cv2.CV_64F, 1, 0, ksize=3) # computing x gradient; kernel size 3
cameraman_sobel_y = cv2.Sobel(cameraman, cv2.CV_64F, 0, 1, ksize=3) # computing y gradient; kernel size 3

cameraman_sobel_magnitude = cv2.magnitude(cameraman_sobel_x, cameraman_sobel_y) # computing the magnitude

max_val = np.max(cameraman_magnitude) # applying the threshold

_, cameraman_threshold10 = cv2.threshold(cameraman_magnitude, max_val*0.1, 255, cv2.THRESH_TOZERO) # 10% of maximum pixel value
_, cameraman_threshold20 = cv2.threshold(cameraman_magnitude, max_val*0.2, 255, cv2.THRESH_TOZERO) # 20% of maximum pixel value

# plotting the figures
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(cameraman_threshold10, cmap='gray')
plt.title("Threshold 10% of maximum pixel value"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cameraman_threshold20, cmap='gray')
plt.title("Threshold 20% of maximum pixel value"), plt.xticks([]), plt.yticks([])
```

3. Non-maximum suppression - check to see if the pixel is a local maximum along the gradient direction. If not, it is set to 0.

**1. Checkerboard image Figure 5(a) (Input image). Change value of thresholdto get detected corners similar to Figure (a) (Harris Corner). Observe and report affect of changing threshold values (3 points)**

```
# function for gaussian 5x5 blurring with sigma = 1. Output is the blurred version of the original image
def gaussian5(image):
    sigma = 1
    kernel = int(5) // 2
    
    x, y = np.mgrid[-kernel:kernel+1, -kernel:kernel+1]
    
    normal = 1 / (2.0 * np.pi * sigma**2)
    g =  np.exp(-((x**2 + y**2) / (2.0*sigma**2))) * normal
    
    return cv2.filter2D(image, -1, g)

# function for finding image derivatives and square of derivatives. Output is H
def gradient(image): 
    sobelx = np.array([[-1, 0, 1], 
                       [-2, 0, 2], 
                       [-1, 0, 1]])
    sobely = np.array([[1, 2, 1], 
                       [0, 0, 0], 
                       [-1, -2, -1]])
    
    image_devX = cv2.filter2D(image, cv2.CV_64F, sobelx)
    image_devY = cv2.filter2D(image, cv2.CV_64F, sobely)
    
    # square of the derivatives 
    image_devXsq = image_devX*image_devX
    image_devYsq = image_devY*image_devY
    image_devXY = image_devX*image_devY
    
    # generating the value of H
    H = np.zeros((image_devX.shape[0], image_devY.shape[1], 2, 2))
    H[:, :, 0, 0] = gaussian5(image_devXsq)
    H[:, :, [0, 1], [1, 0]] = gaussian5((image_devXY))[:, :, np.newaxis]
    H[:, :, 1, 1] = gaussian5(image_devYsq)
    
    return H

# function for non-maxima suppression with a 2x2 window. Output is the image with only the local maxima
def nms(image):
    image_nms = np.zeros_like(image)
    
    # shifting the window in order to identify the maxima
    for i in range(2, image.shape[0]-2): 
        for j in range(2, image.shape[1]-2):
            window = image[i, j]
            if window == 0:
                image_nms[i, j] = 0
            window = image[i-2:i+2+1, j-2:j+2+1]
            if image[i, j] >= np.max(window):                
                image_nms[i, j] = image[i, j]
    
    return image_nms

# function for drawing circles around detected corners. Output is the image with circles drawn on it
def drawCircles(image, f, r):

    points = list(zip(*np.nonzero(f)))
    
    # drawing the points
    for pt in points:
        cv2.circle(image, (pt[1], pt[0]), color = 100, radius=r, thickness=int(r/2))
        
    return image

# combining the previous functions to generate the harris edge detector. Output is the image with detected corners
def harris(image, threshold = 0.1, r=10):  
    image2 = image.copy()
    
    image_gaus = gaussian5(image)
    H = gradient(image_gaus)
    
    trace = H[:, :, 0, 0] + H[:, :, 1, 1] # trace of H
    f = np.linalg.det(H) - 0.05 * np.square(trace) # cornerness function response
    
    f_thresh = f * (f > threshold)
    f_nms = nms(f_thresh)
    
    image_harris = drawCircles(image2, f_nms, r)
    
    return image_harris    

checkerboard_harris = harris(checkerboard, threshold = 0, r = 20)
plt.imshow(checkerboard_harris, cmap = 'gray')
plt.title("Harris Corner Detection Checkerboard"), plt.xticks([]), plt.yticks([])
```

```
small_harris_checkerboard = harris(checkerboard, threshold = 1000, r = 20) # setting small threshold of 1000
medium_harris_checkerboard = harris(checkerboard, threshold = 6000000000, r = 20) # setting medium threshold of 6000000000
large_harris_checkerboard = harris(checkerboard, threshold = 100000000000, r = 20) # setting large threshold of 100000000000

print("threshold = 1000, 6000000000, 100000000000 (from left to right)")
# plotting the figures
fig, axs = plt.subplots(1, 3, figsize=(10, 7), sharey=True)
axs[0].imshow(small_harris_checkerboard, cmap = 'gray')
plt.title("Threshold 1000"), plt.xticks([]), plt.yticks([])
axs[1].imshow(medium_harris_checkerboard, cmap = 'gray')
plt.title("Threshold 6000000000"), plt.xticks([]), plt.yticks([])
axs[2].imshow(large_harris_checkerboard, cmap = 'gray')
plt.title("Threshold 100000000000"), plt.xticks([]), plt.yticks([])
```

Even after implementing large changes in the threshold value, the corner detection as seen in figure 5(a) still occured. At a threshold of 6000000000, some of the corners closer to the border of the checkerboard were not detected. Only at a very high threshold of 100000000000 did the corner detection no longer really work.
